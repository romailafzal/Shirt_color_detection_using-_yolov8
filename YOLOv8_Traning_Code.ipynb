#show predictions

```
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to the main detect folder
detect_folder_path = "/content/runs/detect"

# Get list of all subfolders with names like predict0, predict1, predict2, ...
predict_folders = [folder for folder in os.listdir(detect_folder_path) if folder.startswith("predict")]

# Function to plot images in a folder
def plot_images_in_folder(folder_path):
    image_files = [file for file in os.listdir(folder_path) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        img = mpimg.imread(image_path)
        plt.imshow(img)
        plt.title(image_file)
        plt.axis('off')
        plt.show()

# Loop through each predict folder and plot images
for predict_folder in predict_folders:
    predict_folder_path = os.path.join(detect_folder_path, predict_folder)
    plot_images_in_folder(predict_folder_path)





```
